### 5. Handle the Missing values.

```python
# checking the null values in dataset

data.isna()
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
# if we have null values in numerical column (age) we replace it with mean/median

data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
# if we have null values in categorical column (embarked, deck, embark_town) we replace it with mode

from sklearn.impute import SimpleImputer

impute_mode = SimpleImputer(strategy = 'most_frequent')
impute_mode.fit(data[['embarked', 'deck', 'embark_town']])

data[['embarked', 'deck', 'embark_town']] = impute_mode.transform(data[['embarked', 'deck', 'embark_town']])
```

```python
data.isnull().sum()
```

### 6. Find the outliers and replace the outliers

```python
## outlier detection

sns.boxplot(data.age)
```

```python
perc95=data.age.quantile(0.95)
```

```python
#outlier removal 

data=data[data.age<=perc95]
data = data[data.age>=data.age.quantile(0.1)]
sns.boxplot(data.age)
```

### 7. Check for Categorical columns and perform encoding

```python
## find categorical columns

catCols = data.select_dtypes("object").columns
catCols= list(set(catCols))
catCols
```

```python
## perform encoding (label)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for x in catCols:
    data[x] = le.fit_transform(data[x])
```

```python
data.head()
```

### 8. Split the data into dependent and independent variables

```python
## y - dependent variable

y = data['survived']
y.head()
```

```python
## x - independent variables

x = data.drop(columns=['survived'], axis=1)
x.head()
```

### 9. Scale the independent variables

```python
## min-max scaling

from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()

name = x.columns
x_scaled = scale.fit_transform(x)
x = pd.DataFrame(x_scaled, columns=name)
x.head()
```

### 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=0)
```
